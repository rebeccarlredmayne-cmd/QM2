```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
df_filtered = df[df['TIMESTAMP'].dt.time == pd.to_datetime('12:00:00').time()]
numeric_cols = df_filtered.select_dtypes(include=np.number).columns
for col in numeric_cols:
    df_filtered.loc[df_filtered[col] < -200, col] = np.nan

plt.figure(figsize=(10, 6))
sns.scatterplot(x='TIMESTAMP', y='WS_AVG', data=df_filtered)
plt.title('Average Windspeed per day')
plt.xlabel('Date')
plt.ylabel('Average Windspeed')
plt.grid(True)
plt.show()
```
